```python
images=[]
for file in os.listdir('Voitures'):
    image_path = os.path.join('Voitures', file)
    images.append(image_path)
images
```

```python
List=[]
for image_path in images:
    I=imread(image_path)
    I = color.rgb2gray(I)
    I1 = np.resize(I, (20,20))
    I1= I1.reshape(20*20)
    I1= I1.flatten()
    List.append(I1)
```
