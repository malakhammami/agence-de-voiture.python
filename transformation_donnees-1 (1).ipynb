```python
# importation des librairies
import pandas as pd
import datetime
from sklearn.preprocessing import OneHotEncoder
```

```python
# importation des donn√©es
df  = pd.read_csv("Voitures.csv",sep=',')
df_voitures = df.copy()
df_voitures.head()
```

```python
# suppression de colonne matricule
df_voitures = df_voitures.drop(columns=['matricule'], axis=1)
df_voitures.head()
```

```python
# transformation de la colonne date_circulation en age
df_voitures['date_circulation'] = pd.to_datetime(df_voitures['date_circulation'])
df_voitures
df_voitures['age'] = datetime.datetime.now().year - df_voitures['date_circulation'].dt.year
df_voitures
```

```python
# suppression de la colonne date_circulation
df_voitures = df_voitures.drop(columns=['date_circulation'], axis=1)
df_voitures
```

```python
# transformation de la colonne marque en one hot encoding
OHE = OneHotEncoder()
transformed = OHE.fit_transform(df_voitures[['marque']])
df_voitures.drop(columns = ['marque'],inplace=True)
df_voitures[OHE.categories_[0]] = transformed.toarray()
df_voitures
```

```python
# transformation de la colonne couleur en one hot encoding
OHE = OneHotEncoder()
transformed = OHE.fit_transform(df_voitures[['couleur']])
df_voitures.drop(columns = ['couleur'],inplace=True)
df_voitures[OHE.categories_[0]] = transformed.toarray()
df_voitures
```

```python
#Normaliser les donn√©es
from sklearn.preprocessing import  StandardScaler
ss=StandardScaler()
ss.fit(df_voitures)
X_ss=ss.transform(df_voitures)
X_ss
```

```python
#Reduction des dimensions 
from sklearn.decomposition import  PCA
pca=PCA(n_components=2)
pca.fit(X_ss)
X_pca=pca.transform(X_ss)
X_pca
```
